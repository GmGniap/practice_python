```python
import pandas as pd 
```

```python
player_df = pd.read_csv("player_data.csv")
player_df.head()
```

## Notes 
* isna() function - to look for missing values 
* dropna()  - by default , it removes rows , use 'axis' parameter to remove columns.
    - usually return new DataFrame , use 'inplace' parameter to drop rows/columns in original DataFrame
    - to remove rows/columns in which **all** of values are missing , use 'how' parameter to **all** 
    - 'thresh' parameter - set minimum number of non-NaN values a row/column needs to avoid being dropped



```python
player_df.isna().sum()
player_df.info()
```

```python
# drop columns 
player_df.dropna(axis= 'columns', inplace=True, how='all')
player_df.isna().sum()
```

```python
# drop rows 
player_df.dropna(inplace=True, how='all')
player_df.isna().sum()
```

```python
player_df.shape
```

```python
# Drop all rows that don't have at least 12 non-NaN values 
player_df.dropna(inplace=True, thresh=12)
player_df.isna().sum()
```

```python
player_df.head(10)
```

```python
# reset index after dropping NaN values 
player_df.reset_index(drop=True, inplace=True)
player_df.info()
```

### Outliers
* Outliers are data values so far outside the distribution of other values that they bring into question whether they even belong in the dataset. Outliers often arise from data errors or other undesirable noise. It needs to check for and deal with possible outliers before you analyze the data. 
* Quick way - describe() function
    - Box plots to visualize outliers 

```python
player_df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

- Create list of column names, excluding ID. Use the list to find specific values within each row.
- Create matrix of subplots - to show all 13 column
- Add Padding 
- Create box plot based on data in each column (through loop)

```python
print(player_df.columns)
# create column names list
cols = list(player_df.iloc[:, 1:])
cols

# create 3x5 matrix subplot
fig , axes = plt.subplots(3, 5, figsize=(18,11))

# create padding 
fig.tight_layout(pad=2.0)

# loop over columns of DF and create box plot for each one
for i in range(len(cols)):
    sns.boxplot(ax=axes[i//5, i%5], y=player_df[cols[i]])
```

- idxmin() method returns the index number of the row in which minimum value occurs.

```python
# index -> lowest values in points column 
point_out = player_df['points'].idxmin()
point_out
```

```python
# index -> lowest values in possession 
poss_out = player_df['possessions'].idxmin()
poss_out
```

```python
# drop the row '35' 
player_df.drop(player_df.index[point_out], inplace=True)
player_df.tail(10)
```

```python
# Reset again 
player_df.reset_index(drop=True, inplace=True)
```

```python
# Recheck total of NaN values by row 
player_df.isna().sum()
```
