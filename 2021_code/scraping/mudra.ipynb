```python
import urllib3
from urllib3 import request 
import certifi
import json
import pandas as pd 
```

```python
## Handle Certificate verification 
http = urllib3.PoolManager(cert_reqs = 'CERT_REQUIRED', ca_certs=certifi.where())
```

```python
## Get data from API 
url = 'https://gis2.onemapmyanmar.info/gis2/rest/services/Hosted/Cyclone_wind_Present_query/FeatureServer/0/query?f=json&where=1%3D1&returnGeometry=true&spatialRel=esriSpatialRelIntersects&outFields=objectid%2Cst%2Cst_pcode%2Cdt%2Cdt_pcode%2Cts%2Cts_pcode%2Ct_name_win%2Ct_name_m3%2Cf17_18_fy_m%2Carea_km%2Capartments_house_risk%2Cbamboo_house_risk%2Cbananas_risk%2Cbungalow_brick_house_risk%2Ccoconut_risk%2Chut_house_risk%2Csemi_pacca_house_risk%2Cwooden_house_risk%2Ct_ead_buildings%2Ct_ead_crop%2Ct_ead_alll%2Crisk_economic%2Cdt_name_m3%2Cdt_name_wi%2Cst_name_wi%2Cst_name_m3&outSR=102100&resultOffset=0&resultRecordCount=2000'
r = http.request('GET', url)
print(r.status)
```

```python
## Decode json data into dict object 
data = json.loads(r.data.decode('utf-8'))
```

```python
df = pd.json_normalize(data, 'features')
df.head(10)
```

```python
print(df.columns)
print(df.shape)
```

```python
df.to_csv("./mudra.csv", encoding='utf-8')
```
